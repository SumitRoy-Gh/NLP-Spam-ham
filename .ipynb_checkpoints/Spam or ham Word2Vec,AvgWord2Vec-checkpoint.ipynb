```python
import pandas as pd
import numpy as np
df = pd.read_csv("SMSSpamCollection.txt", sep='\t', names=["label","messages"])
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('stopwords')

```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-Z]',' ',df['messages'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review]
    review=' '.join(review)
    corpus.append(review)
```

```python
print(corpus)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
#simple_preprocess()
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim 
```

```python
#lets train word2vec from scratch
model=gensim.models.Word2Vec(words,vector_size=100)
```

```python
model.wv.index_to_key
```

```python
model.corpus_count
```

```python

def avg_word2vec(doc):
    # remove out-of-vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)
    
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
                #or [np.zeros(len(model.wv.index_to_key))], axis=0)
```

```python
## Dependent Features
## Output Features
y = df[list(map(lambda x: len(x)>0 ,corpus))]
y=pd.get_dummies(y['label'])
y=y.iloc[:,0].values
```

```python
X = []
for i in range(len(words)):
    X.append(avgword2vec(words[i]))

df = pd.DataFrame(X)
```

```python
df.head()
```

```python
df['output']=y
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
X=df.drop('output', axis=1)
```

```python
X.isnull().sum()
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
X_train.head()
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_test,y_pred))
```
